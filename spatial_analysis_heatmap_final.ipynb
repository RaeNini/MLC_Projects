```python
import requests, json

#Search by BBL. example: 4013050036
speed = pd.DataFrame()
speedURL = 'https://data.cityofnewyork.us/resource/i4gi-tjb9.json?$$app_token'

url = requests.get(speedURL)
text = url.text

data = json.loads(text)

```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df['speed'] = df['speed'].apply(pd.to_numeric)
```

```python
arr= df.groupby(['link_points'])['speed'].mean()
```

```python
result = pd.DataFrame(arr, columns=['speed']).reset_index()
result
```

```python
new = result["link_points"].str.split(" ", n = 1, expand = True)
```

```python
result["link"]= new[0]
result
```

```python
new1 = result["link"].str.split(",", n = 1, expand = True)
result["latitude"]= new1[0]
result["longitude"]= new1[1]
result
```

```python
speed_map = folium.Map(location=[40.742459, -73.971765], zoom_start=12)
data = [[x[0], x[1], 1] for x in np.array(result[['latitude', 'longitude']])]
HeatMap(data, radius = 20).add_to(speed_map)
speed_map
```
